```python
df=df[['Close']]
df.head(5)
```

```python
future_days= 25
df['prediction']=df[['Close']].shift(-future_days)
df.head(6)
```

```python
x=np.array(df.drop(['prediction'],1))[:-future_days]
print(x)
```

```python
y=np.array(df['prediction'])[:-future_days]
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
tree=DecisionTreeRegressor().fit(x_train,y_train)
lr=LinearRegression().fit(x_train,y_train)
```

```python
x_future=df.drop(['prediction'],1)[:-future_days]
x_future=x_future.tail(future_days)
x_future=np.array(x_future)
x_future
```

```python
tree_prediction=tree.predict(x_future)
print(tree_prediction)
print()
lr_prediction=lr.predict(x_future)
print(lr_prediction)
```

```python
prediction=tree_prediction
valid=df[x.shape[0]:]
valid['prediction']=prediction
plt.figure(figsize=(16,8))
plt.title('Model')
plt.xlabel('Days')
plt.ylabel('Close price')
plt.plot(df['Close'])
plt.plot(valid[['Close','prediction']])
plt.legend(['orig','val','pred'])
plt.show()
```

```python
def create_essential_data():
    essential_items = ["Grocery", "Home Supplies", "Toiletries", "Medicine"]
    essential_data = {}

    for item in essential_items:
        value = input(f"Enter the amount for {item}: ")
        essential_data[item] = value

    return essential_data

def create_people_data(num_people):
    people_data = {}

    for i in range(num_people):
        name = input(f"Enter the name for Person {i+1}: ")
        people_data[f"Person {i+1}"] = name

    return people_data

def create_monthly_data():
    monthly_items = ["Light Bill", "Education", "TV", "Mobile", "Home Rent", "Gas", "Others"]
    monthly_data = {}

    for item in monthly_items:
        value = input(f"Enter the amount for {item}: ")
        monthly_data[item] = value

    return monthly_data

def main():
    print("Expense Tracker\n")

    essential_data = create_essential_data()

    num_people = int(input("Enter the total number of people in home: "))
    people_data = create_people_data(num_people)

    monthly_data = create_monthly_data()

    print("\nExpense Summary:")
    print("Essential Items:")
    for item, value in essential_data.items():
        print(f"{item}: {value}")

    print("\nPeople in Home:")
    for person, name in people_data.items():
        print(f"{person}: {name}")

    print("\nMonthly Expenses:")
    for item, value in monthly_data.items():
        print(f"{item}: {value}")

if __name__ == "__main__":
5    main()
```
